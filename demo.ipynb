```python
import yfinance as yf
import matplotlib.pyplot as plt
import datetime
from datetime import date,timedelta
import pandas as pd
```

```python
df_FORD = yf.Ticker("F").history(period="6y")
```

```python
df_FORD
```

```python
date.today()
```

```python

# Data extracted for range of days.
DATA_RANGE_DAYS = 365*5

# Get the dates in datetime format for extracting the details
END_DATE = date.today()
START_DATE = END_DATE - timedelta(DATA_RANGE_DAYS)
print(f'Data:{START_DATE} to {END_DATE}')
```

```python
plt.figure(figsize=(10,8))
plt.title('Close Price History')
plt.xlabel('Date', fontweight = 'bold')
plt.ylabel('Close Price USD ($)',fontweight = 'bold')
plt.plot(df_FORD['Close'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title('Volume History')
plt.xlabel('Volume', fontweight = 'bold')
plt.ylabel('Volume Price USD ($)',fontweight = 'bold')
plt.plot(df_FORD['Volume'])
plt.show()
```

### CANDLE STICK GRAPH

```python
up = df_FORD[df_FORD['Close'] >= df_FORD['Open']]
down = df_FORD[df_FORD['Close'] < df_FORD['Open']]

#If the stock prices have decreased then the candlestick will be red in color
col1= 'red'

#If the stock prices have increased then the candlestick will be green in color
col2 = 'green'

width = 20
width2 = 7.5

plt.figure(figsize=(16,10))
# Plotting up prices of the stock 
plt.title('Candlestick Graph', fontweight='bold')
plt.bar(up.index, up['Close']-up['Open'], width, bottom=up['Open'], color=col1) 
plt.bar(up.index, up['High']-up['Close'], width2, bottom=up['Close'], color=col1) 
plt.bar(up.index, up['Low']-up['Open'], width2, bottom=up['Open'], color=col1) 
  
# Plotting down prices of the stock 
plt.bar(down.index, down['Close']-down['Open'], width, bottom=down['Open'], color=col2) 
plt.bar(down.index, down['High']-down['Open'], width2, bottom=down['Open'], color=col2) 
plt.bar(down.index, down['Low']-down['Close'], width2, bottom=down['Close'], color=col2) 
```
